````python
# code here
prob = 0.5
start = 2
positions = [start]
print(positions)
steps=100
rr = np.random.random(1000)

for i in range(steps):
  if rr[i] < prob:
    positions.append(positions[i-1] + 1)
  else:
    positions.append(positions[i-1] -1)

plt.plot(positions)
plt.show()
````

````python
s = 100000000

x = np.zeros(s)
y = np.zeros(s)

for i in range(1,s):
  val = random.randint(1,4)
  if val == 1:
    x[i] = x[i-1] + 1
    y[i] = y[i-1]
  elif val == 2:
    x[i] = x[i-1] - 1
    y[i] = y[i-1]
  elif val == 3:
    x[i] = x[i-1]
    y[i] = y[i-1] + 1
  else:
    x[i] = x[i-1]
    y[i] = y[i-1] - 1

#plot
plt.plot(x,y)
````
